```python
# Make a class called Employee. The __init__() method for Employee should store its different attributes:
# name, number, address, salary, post, year of joining, education, and his year of experience.

# Make a method called describe_employee() that prints the detail Information about employee,
# a method update_ year_of_experience(), and a method called increase_salary() that do increamnet in it's salary.

# Make an instance called employee101 from your class. Print the four attri-
# butes individually, and then call all it's methods for describing employee, ssalary increamnet and increase year of experience.

# Create 10 different instances from the class, and call describe_employee() for each
# instance. Then do some updations with it's methods.

# Finally complete the same task without using the Classes or OOP.


class Employee:
    def __init__(self, name, number, address, salary, post, year_of_joining, education, year_of_experience):
        self.name = name
        self.number = number
        self.address = address
        self.salary = salary
        self.post = post
        self.year_of_joining = year_of_joining
        self.education = education
        self.year_of_experience = year_of_experience

    def describe_employee(self):
        print(f"Name: {self.name}")
        print(f"Serial No.: {self.number}")
        print(f"Address: {self.address}")
        print(f"Salary: Rs.{self.salary}")
        print(f"Post: {self.post}")
        print(f"Year of Joining: {self.year_of_joining}")
        print(f"Education: {self.education}")
        print(f"Years of Experience: {self.year_of_experience}")

    def update_year_of_experience(self, additional_years):
        self.year_of_experience += additional_years
        print(f"Updated Years of Experience for {self.name}: {self.year_of_experience}")

    def increase_salary(self, increment):
        self.salary += increment
        print(f"Updated Salary for {self.name}: Rs.{self.salary}")
```

```python
# Creating an instance of Employee
employee101 = Employee(
    name="Anas",
    number="EMP101",
    address="C-70 block 11",
    salary=50000,
    post="Data Analytics",
    year_of_joining=2022,
    education="Bachelor's in Computer Science",
    year_of_experience=2
)

# Printing the four attributes individually
print("four attributes individually are: ")
print(employee101.name)
print(employee101.number)
print(employee101.address)
print(employee101.salary)
print()

# Calling methods to describe the employee, increase salary, and update experience
print("Describe Employee in detailed: ")
employee101.describe_employee()
employee101.increase_salary(5000)
employee101.update_year_of_experience(1)
```
